```python
product_price = []
cust_names = []
cust_comments = []
cust_stars = []
cust_reviews = []
from bs4 import BeautifulSoup as bs
import requests
r = requests.get(url)
r.content
htmlcontent=r.content
htmlcontent
soup=bs(htmlcontent,'html.parser')
print(soup)
print(soup.prettify())
soup.title
title = soup.title
print(title.string)
price = soup.find_all('div',class_="_30jeq3 _16Jk6d")
price
for i in range(0,len(price)):
    product_price.append(price[i].get_text())
names = soup.find_all('p',class_='_2sc7ZR _2V5EHH')
for i in range(0,len(names)):
    cust_names.append(names[i].get_text())
comments = soup.find_all('p',class_='_2-N8zT')
for i in range(0,len(comments)):
    cust_comments.append(comments[i].get_text())
star = soup.find_all('div',class_='_3LWZlK _1BLPMq')
for i in range(0,len(star)):
    cust_stars.append(star[i].get_text())
review = soup.find_all('div',class_='t-ZTKy')
for i in range(0,len(review)):
    cust_reviews.append(review[i].get_text())        
```
